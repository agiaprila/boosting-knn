```python
#packages
import pandas as pd
```

```python
#panggil data
df = pd.read_excel('data status gizi balita.xlsx')
df
```

```python
df.groupby('status gizi').size()
```

```python
# buat golongan data frame
baik = df[df['status gizi']=='baik']
BGM = df[df['status gizi']=='BGM']
kurang = df[df['status gizi']=='kurang']
```

```python
baik.describe()
```

```python
BGM.describe()
```

```python
kurang.describe()
```

```python
# count menjelaskan bahwa semua 4 features mempunyai 150 row
# secara umum, dari rata-rata kita dapat mengatakan bahwa sepal lebih besar dari petal
import matplotlib.pyplot as plt
plt.figure()
fig,ax=plt.subplots(1,2,figsize=(15, 5))
#umur dan berat
baik.plot(x="umur (bln)", y="berat (kg)", kind="scatter",ax=ax[0],label='baik',color='r')
kurang.plot(x="umur (bln)",y="berat (kg)",kind="scatter",ax=ax[0],label='kurang',color='b')
BGM.plot(x="umur (bln)", y="berat (kg)", kind="scatter", ax=ax[0], label='BGM', color='g')

#tinggi dan lika
baik.plot(x="tinggi (cm)", y="lika (cm)", kind="scatter",ax=ax[1],label='baik',color='r')
kurang.plot(x="tinggi (cm)",y="lika (cm)",kind="scatter",ax=ax[1],label='kurang',color='b')
BGM.plot(x="tinggi (cm)", y="lika (cm)", kind="scatter", ax=ax[1], label='BGM', color='g')

ax[0].set(title='Perbandingan Umur dan Berat', ylabel='Berat Badan')
ax[1].set(title='Perbandingan Tinggi dan Lingkaran Kepala',  ylabel='Lebar Kepala')
ax[0].legend()
ax[1].legend()

```
